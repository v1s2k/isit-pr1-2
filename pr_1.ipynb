```python
from apriori_python import apriori
from efficient_apriori import apriori as eff_apriori
import csv
from pprint import pprint
```

```python
with open("assoc.csv", "r", encoding="utf-8") as f:
    data = list(csv.reader(f))

objects = data[0]

def map_transaction(norm, objects):
    t = []
    for i, value in enumerate(norm):
        if int(value):
            t.append(objects[i])
    return t

transations = [map_transaction(x, objects) for x in data[1:]]
print(*transations, sep="\n")
```

```python
def get_collection_freq(*col):
    freq = 0
    for i in transations:
        if set(col) <= set(i):
            freq += 1
    return freq


```

```python
for i, obj_a in enumerate(objects):
    for j, obj_b in enumerate(objects):
        if j <= i :
            continue
        freq = get_collection_freq(obj_a, obj_b)
        if freq>=4:
            print(obj_a, obj_b, freq)
```

```python
for i, obj_a in enumerate(objects):
    for j, obj_b in enumerate(objects):
        for k, obj_c in enumerate(objects):
            if j <= i or k <= j:
                continue
            freq = get_collection_freq(obj_a, obj_b, obj_c)
            if freq>=4:
                print(obj_a, obj_b,obj_c, freq)
```

```python
for i, obj_a in enumerate(objects):
    for j, obj_b in enumerate(objects):
        for k, obj_c in enumerate(objects):
            for l, obj_d in enumerate(objects):
                if j <= i or k <= j or l<=k:
                    continue
                freq = get_collection_freq(obj_a, obj_b, obj_c, obj_d)
                if freq>=4:
                    print(obj_a, obj_b,obj_c, obj_d, freq)
```
